```python
import torch
import torch.nn as nn
```

```python
class LeNet(nn.Module):
    def __init__(self):
        super(LeNet, self).__init__()
        self.relu = nn.ReLU()
        # removed kernel_size and stride arguments. Added output_size argument.
        self.pool = nn.AdaptiveAvgPool2d(output_size=(2,2))
        self.conve1 = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=(2,2), stride=(1,1), padding=(0,0))
        # reduced kernel_size to (3,3) from (5,5)
        self.conve2 = nn.Conv2d(in_channels=6, out_channels=16, kernel_size=(2,2), stride=(1,1), padding=(0,0))
        # reduced kernel_size to (1,1) from (3,3) to avoid the error
        self.conve3 = nn.Conv2d(in_channels=16, out_channels=120, kernel_size=(1,1), stride=(1,1), padding=(0,0))
        # changed in_features to 480 from 120
        self.linear1 = nn.Linear(in_features=480, out_features=84)
        self.linear2 = nn.Linear(in_features=84, out_features=10)

    def forward(self, x):
      x = self.relu(self.conve1(x))
      x = self.pool(x)
      x = self.relu(self.conve2(x))
      x = self.pool(x)
      x = self.relu(self.conve3(x))
      x = x.reshape(x.shape[0], -1)
      x = self.relu(self.linear1(x))
      x = self.linear2(x)
      return x
```

```python
x = torch.randn(64, 1, 32, 32)
model = LeNet()
print(model(x).shape)
```
